```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/edustats.csv')
train = train.dropna()
train[:10]
```

```python
train.shape
```

```python
train = train.sort_values('value',ascending=False)
train[:5]
```

```python
train['rank'] = train.groupby('Indicator.Code').cumcount()+1
train[:5]
```

```python
train = train.pivot(index='Country.Name',columns='Indicator.Code',values='rank').reset_index() # case1
train = train.pivot(index='Country.Name',columns='Indicator.Code',values='value').reset_index() # case2
train
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer()
# Found that using this imputer wasnt effective
```

```python
rank_df = pd.read_csv('data/imo.csv')
rank_df[:5]
```

```python
train = pd.merge(train,rank_df,on='Country.Name')
train.shape
```

```python
train_columns = train.columns[1:-1]
train_columns
```

```python
for col in train_columns:
    if train[col].corr(train['Mean.Rank'])>=0:
        train[col]=train[col].fillna(train[col].max())
    else:
        train[col] = train[col].fillna(train[col].min())
```
